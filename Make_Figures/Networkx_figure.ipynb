```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G=nx.Graph()
G.add_node(1,pos=(0,1))
G.add_node(2,pos=(1,1))
G.add_node(3,pos=(1,0))
G.add_edge(1,2,weight=0.08)
G.add_edge(2,3,weight=0.06)
G.add_edge(1,3,weight=0.01)
pos=nx.get_node_attributes(G,'pos')
nx.draw(G,pos,node_size = 1000, width=10, with_labels= True, font_weight='bold')
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.savefig('/Users/bengieru/Desktop/mln_v0/network.pdf')
```
