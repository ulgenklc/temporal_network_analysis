```python
z_data_MMM = pd.read_csv('/Users/bengieru/MLN/Make_Figures/MMM_landscape.txt', delimiter = ",")

z = z_data_MMM.values
sh_0, sh_1 = z.shape
x, y = np.linspace(0, 1.5, sh_0), np.linspace(0, 1.5, sh_1)
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='MMM landscape', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))


fig.update_traces(colorscale='Edge', selector=dict(type='surface')) 
fig.update_traces(cmax=1, selector=dict(type='surface'))
fig.update_traces(cmin=0, selector=dict(type='surface'))
fig.show()

py.iplot(fig, filename='MMM')
```
